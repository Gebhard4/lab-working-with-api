```python
def get_codes():
    
    DexNo = input('Insert DexNo:')
    print()
    
    url = "https://poke-info-api.p.rapidapi.com/pokemons/dexNo"
    querystring = {"dexNo":DexNo.lower(), "limit":"10"}

    response = requests.get(url, headers=headers, params=querystring)
    codes = response.json()

    return codes


def choose_code():
    
    codes = get_codes()
    #print(codes)
    
    while type(codes) is str or len(codes) > 1:
        
        if not codes:
            print('Try another DexNo. Unable to find number for:', codes, '\n')
        else:
            print('More than one Pokemon found. Insert one of the following Pokemon:', [code['shortName'] for code in codes], '\n')
        
        codes = get_codes()
    
    return codes['dexNo']
```
